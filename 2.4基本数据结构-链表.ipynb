```python
# 2.4 链表以及数组的实现

class Node:
    def __init__(self, initdata):
        self.data = initdata
        self.next = None
        
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
    
    def setData(self, newdata):
        self.data = newdata
        
    def setNext(self, newnext):
        self.next = newnext
```

```python
# 无序数组

'''
UnorderedList():创建一个新的空链表
add(item):向列表中添加元素
remove(item):从列表中删除该项
search(item):在列表中搜索该项
isEmpty():检查列表是否为空
size():返回列表中项数
append(item):将项添加到列表末尾
index(item):返回项在列表中的位置
insert(pos, item):在制定位置处加入项
pop():删除并返回最后一项
pop(pos):删除并返回指定位置的项
'''

class UnorderedList:
    def __init__(self):
        self.head = None
        
    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
    
    def append(self, item):
        current = self.head
        if current == None:
            self.head = Node(item)
        else:
            while current.getNext() != None:
                current = current.getNext()
            current.setNext(Node(item))
    
    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current == None:
                raise NameError("Data not found in the list")
            elif current.getData() == item:
                found = True
            else:
                previous = current
                current = current.getNext()
        if previous == None:
            self.head = current.getNext()
        else:
            previous.setNext(current.getNext())
        
    def insert(self, pos, item):
        if pos == 0:
            head = self.head
            self.head = Node(item)
            self.head.setNext(head)
        else:
            current = self.head
            for i in range(pos-1):
                if current == None:
                    raise IndexError("Invalid index")
                current = current.getNext()
            next = current.getNext()
            current.setNext(Node(item))
            if next != None:
                current = current.getNext()
                current.setNext(next)
    
    def pop1(self):
        current = self.head
        next = current.getNext()
        if current == None:
            raise IndexError("No data available in the list")
        elif next == None:
            self.head = None
            return current.getData()
        else:
            while next != None:
                previous = current
                current = current.getNext()
                next = current.getNext()
            previous.setNext(None)
            return current.getData()
    
    def pop2(self, pos):
        current = self.head
        if current == None:
            raise IndexError("No data available in the list!")
        if pos == 0:
            self.head = current.getNext()
            return current.getData()
        previous = current
        current = current.getNext()      
        for i in range(pos-1):
            if current == None:
                raise IndexError("Invalid index!")
            else:
                previous = current
                current = current.getNext()
        previous.setNext(current.getNext())
        return current.getData()
    
    def pop(self, pos=None):
        if pos == None:
            self.pop1()
        else:
            self.pop2(pos)
        
    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()
        return found
        
    def isEmpty(self):
        return self.head == None
    
    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count
    
    def index(self, item):
        current = self.head
        id = 0
        found = False
        while not found:
            if current == None:
                raise NameError("Data not found in the list")
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()
                id = id + 1
        return id 
    
    def show(self):
        current = self.head
        while current != None:
            print(current.getData(), end="")
            if current.getNext() != None:
                print(" -> ", end="")
            current = current.getNext()
        print('\n')
```

```python
u = UnorderedList()
u.add(12)
u.add(24)
u.add(35)
u.show()
u.remove(24)
u.show()
print(u.search(3))
print(u.isEmpty())
print(u.size())
u.append(33)
# u.index(3)
u.show()
u.insert(0, 100)
u.show()
u.pop()
u.show()
u.pop(1)
u.show()
```
